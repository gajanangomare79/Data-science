# Hyperparameter for RandomForestClassifier

```python
rfc1=RandomForestClassifier(random_state=42)
param_grid = { 

     'max_features': ['sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
model32= RandomizedSearchCV(estimator=rfc1,param_distributions=param_grid ,cv= 5)
model32.fit(x_train,y_train)
```

# Staking model

```python
models = [ RandomForestClassifier(), MLPClassifier(), LinearSVC(),  DecisionTreeClassifier() ]
      
S_Train, S_Test = stacking(models,                   
                           x_train, y_train, x_test,   
                           regression=False, 
     
                           mode='oof_pred_bag', 
       
                           needs_proba=False,
         
                           save_dir=None, 
            
                           metric=accuracy_score, 
    
                           n_folds=4, 
                 
                           stratified=True,
            
                           shuffle=True,  
            
                           random_state=0,    
         
                           verbose=2)
```

# Meta_classifier as KNN

```python
metamodel=KNeighborsClassifier()
metamodel.fit(S_Train,y_train)
final_predict=metamodel.predict(S_Test)
print(classification_report(final_predict,y_test))
```
