```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
Semi_para=pd.DataFrame(data=Naive_dataset, columns=["BloodPressure","BMI"])
```

```python
X=Semi_para
y=Naive_dataset['Outcome']
```

```python
n_neighbors = 5
cmap_light = ListedColormap(["orange", "cyan", "cornflowerblue"])
cmap_bold = ["darkorange", "c", "darkblue"]


for weights in ["uniform", "distance"]:
    # we create an instance of Neighbours Classifier and fit the data.
    model = neighbors.KNeighborsClassifier(n_neighbors, weights=weights)
    model.fit(X, y)

    _, ax= plt.subplots()
    DecisionBoundaryDisplay.from_estimator(model,X,cmap=cmap_light,ax=ax,response_method="predict",plot_method="pcolormesh",
        xlabel=X.columns[0],
        ylabel=X.columns[1],
        shading="auto")

 # Plot also the training points'

    sns.scatterplot(
        x=X.iloc[:,0],
        y=X.iloc[:,1],
        hue=y,
        palette=cmap_bold,
        alpha=1.0,
        edgecolor="black",
       )
    
    plt.title(
        "3-Class classification (k = %i, weights = '%s')" % (n_neighbors, weights));
```

```python
HousePrice=pd.read_csv(r"C:\Users\ggomare\Downloads\housePrice-2.csv")
```

```python
HousePrice
```

```python
X1_train=HousePrice.drop("Price (USD)",axis=1)
```

```python
X1_train.shape
```

```python
Y1_test=HousePrice["Price (USD)"]
```

```python
Y1_test
```

```python
X1train, X1test, y1train, y1test=train_test_split(X1_train,Y1_test,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler 
 
scaler = StandardScaler() 
# Fit only on X_train 
scaler.fit(X1train) 
 
# Scale both X_train and X_test 
X1train = scaler.transform(X1train) 
X1test = scaler.transform(X1test)
```

```python
X1train
```

```python
from sklearn.neighbors import KNeighborsRegressor    
 
regressor = KNeighborsRegressor(n_neighbors=5) 
 
regremodel=regressor.fit(X1train, y1train)
```

```python
output5=regremodel.predict(X1test)
```

```python
output5
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error    
 
mae = mean_absolute_error(y1test, output5) 
mse = mean_squared_error(y1test, output5)
 
print('mae is: ', mae) 
print('mse is: ', mse) 
```

```python
R2= regressor.score(X1test, y1test) 
 
print('R2 is:', R2) 
```

```python
error=[]
for i in range(1, 30):   
    knn = KNeighborsRegressor(n_neighbors=i) 
    knn.fit(X1train, y1train) 
    pred_i = knn.predict(X1test) 
    mae = mean_absolute_error(y1test, pred_i) 
    error.append(mae)
```

```python
error
```

```python
import matplotlib.pyplot as plt  
 
plt.figure(figsize=(12, 6))    
 
plt.plot(range(1, 30), error, color='red', linestyle='dashed', marker='o',markerfacecolor='blue', markersize=10) 
          
plt.title('K Value MAE') 
plt.xlabel('K Value') 
plt.ylabel('Mean Absolute Error (MAE)');
```

```python
error2 = [] 
 
# Calculating MAE error for K values between 1 and 39
for i in range(1, 30): 
    knn = KNeighborsRegressor(n_neighbors=i) 
    knn.fit(X1train, y1train) 
    pred_i = knn.predict(X1test) 
    mse = mean_squared_error(y1test, pred_i) 
    error2.append(mse) 
```

```python
import matplotlib.pyplot as plt  
 
plt.figure(figsize=(12, 6)) 
 
plt.plot(range(1, 30), error2, color='green', linestyle='dashed', marker='o',markerfacecolor='red', markersize=10) 
          
plt.title('K Value MSE') 
plt.xlabel('K Value') 
plt.ylabel('Mean Squared Error (MSE)');
```
