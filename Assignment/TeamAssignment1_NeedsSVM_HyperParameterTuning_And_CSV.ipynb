```python
trainfile = r"C:\Users\ggomare\Documents\pre1\train.csv"
trainData = pd.read_csv(trainfile) #creates a dataframe

testfile = r"C:\Users\ggomare\Documents\pre1\test.csv"
testData = pd.read_csv(testfile)  #creates a dataframe


print(trainData.shape)
print(testData.shape)
```

```python
#Drop ID, OpenDate, and City collumns from Test and Train Data 
test_data = testData.drop(columns=['Id', 'Open Date', 'City','Type'], axis = 1)

train_data = trainData.drop(columns=['Id', 'Open Date', 'City','Type'], axis = 1)

print(train_data.shape)
print(test_data.shape)


train_data["City Group"].replace({
    'Big Cities':0,
    'Other':1
},inplace=True)

test_data["City Group"].replace({
    'Big Cities':0,
    'Other':1
},inplace=True)

```

```python
test_data
```

```python
#Extract training and test data
ytrain = train_data[['revenue']].copy()
Xtrain = train_data.drop(["revenue"], axis=1) #extracting training data without the target column

print(Xtrain.shape)
print(ytrain.shape)
Xtest = test_data.copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xtrain, ytrain, test_size = .30, random_state = 51)
```

# Original Decision Tree Model With Hyperparameter Tuning

```python
#Decision Tree Regressor ========================================================================
#CONSTRUCT DEFAULT DECISION TREE AND OBTAIN RESPECTIVE ACCURACY 
clf = DecisionTreeRegressor()
clf.fit(X_train, y_train)
clf_predict_Train=clf.predict(X_train)


mean_squared_error(y_train,clf_predict_Train)
print("RMSE (training) for Decision Tree:{0:10f}".format(mean_squared_error(y_train,clf_predict_Train)))
clf_predict_Test=clf.predict(X_test)
mean_squared_error(y_test,clf_predict_Test)
print("RMSE (Test Data) for Decision Tree:{0:10f}".format(mean_squared_error(y_test,clf_predict_Test)))
```

```python
#Hyperparameter tuning done for decision tree regressor
parameters={'min_samples_split' : range(10,100,10),'max_depth': range(1,20,2)}
clf_random = RandomizedSearchCV(clf,parameters,n_iter=15)
clf_random.fit(X_train, y_train)
grid_parm=clf_random.best_params_
print(grid_parm)

#Using the parameters obtained from HyperParameterTuning in the DecisionTreeClassifier 
clf2 = DecisionTreeRegressor(**grid_parm)
clf2.fit(X_train,y_train)
clf_predict = clf2.predict(X_test)
clf_predictDT = pd.DataFrame(clf_predict,columns=['Revenue'])
```

# Cross-Validation Score

```python
clf_cv_score = cross_val_score(clf2, Xtrain, ytrain, cv=5)
clf_cv_score
```

```python
pred1=clf2.predict(test_data)
id=testData["Id"]
final1=pd.DataFrame(pred1,columns=["Prediction"])
resul1=pd.concat([id,final1],axis=1)
resul1.to_csv("submission1.csv",index=False)

resul1.head()
```

```python
#STACKING MODELS ===================================================================== we do this but with other ones like SVM neural net etc
print("___________________________________________________________________________________________\nEnsemble Methods Predictions using GradientBoosting, RandomForest and Decision Tree Classifier\n")

models = [ LinearSVR(), RandomForestRegressor(), DecisionTreeRegressor() ]
      
S_Train, S_Test = stacking(models,                   
                           Xtrain, ytrain, test_data,#xtest needs to pass through the base models  
                           regression=True, 
     
                           mode='oof_pred_bag', 
       
                           needs_proba=False,
         
                           save_dir=None, 
                                        
                           n_folds=4,
                           shuffle = True,
                                                    
                           verbose=2)

```

# Stacked Decision Tree Hyper Parameter Tuning 

```python
modelDT = DecisionTreeRegressor()
    
modelDT = modelDT.fit(S_Train, ytrain)
y_pred = modelDT.predict(S_Test)
Y_PredDT = pd.DataFrame(y_pred,columns=['Revenue'])

```

```python
#Hyperparameter tuning done for decision tree regressor
parameters={'min_samples_split' : range(10,100,10),'max_depth': range(1,20,2)}
clf_random = RandomizedSearchCV(modelDT,parameters,n_iter=15)
clf_random.fit(S_Train, ytrain)
grid_parm=clf_random.best_params_
print(grid_parm)

#Using the parameters obtained from HyperParameterTuning in the DecisionTreeClassifier 
clf = DecisionTreeRegressor(**grid_parm)
clf.fit(S_Train,ytrain)
clf_predict = clf.predict(S_Test)
clf_predictDT = pd.DataFrame(clf_predict,columns=['Revenue'])



#get cross-validation report
clf_cv_score = cross_val_score(clf, S_Train, ytrain, cv=10, scoring="roc_auc")

```

```python
clf_predict = clf.predict(S_Test)
id=testData["Id"]
final_Stacked_DT=pd.DataFrame(clf_predict,columns=["Prediction"])
results_Stacked_DT=pd.concat([id,final_Stacked_DT],axis=1)
results_Stacked_DT.to_csv("Stacked_Dt_Results.csv",index=False)
```
