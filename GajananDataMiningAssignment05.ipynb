# Hyperparameter tunning

```python
Question1 = []
parameter1=['k-means++', 'random']
parameter2=[2,4,6,8,10,12]
k_rng = range(1,20)
for k in k_rng:
    for P in parameter1:
        for Q in parameter2:
            means= KMeans(n_clusters=k,init=P,n_init=Q)
            means.fit(dataset1)
            Question1.append(means.inertia_)
```

```python
sse = []
k_rng = range(1,20)
for k in k_rng:
    means = KMeans(n_clusters=k)
    means.fit(dataset1)
    sse.append(means.inertia_)
```

```python
from matplotlib import pyplot as plt
plt.plot(k_rng,sse)
```

```python
means=KMeans(n_clusters=3)
predicted=means.fit_predict(dataset1)
predicted
```

```python
from collections import Counter
Counter(predicted)
```

```python
dataset1["predicted"]=predicted
```

```python
label_0=dataset1[dataset1["predicted"]==0]
label_1=dataset1[dataset1["predicted"]==1]
label_2=dataset1[dataset1["predicted"]==2]
```

# Hyperparameter Tunning

```python
import warnings
warnings.filterwarnings("ignore")
Question2 = []
parameter1=['k-means++', 'random']
parameter2=[2,4,6,8,10,12]
k_rng = range(1,20)
for k in k_rng:
    for l in parameter1:
        for m in parameter2:
            cluster_model= KMeans(n_clusters=k,init=l,n_init=m)
            cluster_model.fit(dataset21)
            Question2.append(cluster_model.inertia_)
```

```python
question21 = []
k_rng = range(1,20)
for k in k_rng:
    cluster_model = KMeans(n_clusters=k)
    cluster_model.fit(dataset21)
    question21.append(cluster_model.inertia_)
```
