```python
dataset1=dataset[(dataset.Cholesterol < dataset.Cholesterol.mean()+3*dataset.Cholesterol.std())]
```

```python
dataset1.describe()
```

```python
dataset1.MaxHR.mean()+3*dataset1.MaxHR.std()
```

```python
dataset2=dataset1[dataset1.MaxHR < dataset1.MaxHR.mean()+3*dataset1.MaxHR.std()]
```

```python
dataset2
```

```python
dataset3=dataset2[dataset2.RestingBP < dataset2.RestingBP.mean()+3*dataset2.RestingBP.std()]
```

```python
dataset3
```

```python
dataset4=dataset3[dataset3.Oldpeak < dataset3.Oldpeak.mean()+3*dataset3.Oldpeak.std()]
```

```python
dataset4
```

```python
dataset4.RestingECG.unique()
```

```python
dataset5 = dataset4.copy()
dataset5.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

dataset5.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

dataset5.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

dataset5.head()
```

```python
dataset5=pd.get_dummies(dataset5,drop_first=True)
```

```python
dataset5
```

```python
x=dataset5.drop(columns=["HeartDisease"])
y=dataset5["HeartDisease"]
```

```python
x
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1=LogisticRegression()
```

```python
model1.fit(xtrain,ytrain)
```

```python
model1.score(xtest,ytest)
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(LogisticRegression(),x,y,cv=5)
```

```python
score.mean()
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
bagging_model=BaggingClassifier(base_estimator=DecisionTreeClassifier(),
                               n_estimators=100,
                               max_samples=0.8,
                               oob_score=True,
                               random_state=0)
bagging_model.fit(x,y)
```
