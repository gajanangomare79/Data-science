```python
import pandas as pd
import numpy as np
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
from matplotlib import pyplot as plt
```

```python
dataset=pd.read_csv(r"C:\Users\ggomare\Downloads\income.csv")
```

```python
dataset
```

```python
plt.scatter(dataset.Age,dataset["Income($)"])
```

```python
mean=KMeans(n_clusters=3)
pred=mean.fit_predict(dataset[["Income($)","Age"]])
```

```python
dataset["label"]=pred
```

```python
df0=dataset[dataset["label"]==0]
df1=dataset[dataset["label"]==1]
df2=dataset[dataset["label"]==2]
```

```python
plt.scatter(df0["Age"],df0["Income($)"],color='red')
plt.scatter(df1["Age"],df1["Income($)"],color='green')
plt.scatter(df2["Age"],df2["Income($)"],color='black')
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(dataset[['Income($)']])
dataset['Income($)'] = scaler.transform(dataset[['Income($)']])
scaler.fit(dataset[['Age']])
dataset['Age'] = scaler.transform(dataset[['Age']])
```

```python
dataset
```

```python
plt.scatter(dataset["Age"],dataset["Income($)"])
```

```python
pred2=mean.fit_predict(dataset[["Age","Income($)"]])
```

```python
dataset["new_label"]=pred2
```

```python
dataset
```

```python
dfN0=dataset[dataset["new_label"]==0]
dfN1=dataset[dataset["new_label"]==1]
dfN2=dataset[dataset["new_label"]==2]
```

```python
plt.scatter(dfN0["Age"],dfN0["Income($)"],color='red')
plt.scatter(dfN1["Age"],dfN1["Income($)"],color='green')
plt.scatter(dfN2["Age"],dfN2["Income($)"],color='black')
```
