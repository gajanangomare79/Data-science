```python
use_column=["Suburb","Rooms","Type","Method","SellerG",
            "Regionname","Propertycount","Distance","CouncilArea",
           "Bedroom2","Bathroom","Car","Landsize","BuildingArea","Price"]
```

```python
dataset1=dataset[use_column]
```

```python
colu_fillna=["Bedroom2","Car","Propertycount","Bathroom","Distance"]
```

```python
dataset1[colu_fillna]=dataset1[colu_fillna].fillna(0)
```

```python
dataset1.isnull().sum()
```

```python
dataset1["Landsize"]=dataset1["Landsize"].fillna(dataset1.Landsize.mean())
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset1["BuildingArea"]=dataset1["BuildingArea"].fillna(dataset1.BuildingArea.mean())
```

```python
dataset1["Price"]=dataset1["Price"].fillna(dataset1.Price.mean())
```

```python
dataset1
```

```python
dataset1["SellerG"].unique()
```

```python
dataset2=pd.get_dummies(dataset1,drop_first=True)
```

```python
dataset2.shape
```

```python
dataset2
```

```python
dataset2.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
x=dataset2.drop("Price",axis=1)
y=dataset2["Price"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20)
```

```python
model1=LinearRegression()
```

```python
model1.fit(xtrain,ytrain)
```

```python
model1.score(xtest,ytest)
```

```python
model1.score(xtrain,ytrain)
```

```python
model2=Lasso(alpha=50,max_iter=100,tol=0.1)
```

```python
model2.fit(xtrain,ytrain)
```

```python
model2.score(xtrain,ytrain)
```

```python
model2.score(xtest,ytest)
```

```python
model3=Ridge(alpha=50,max_iter=100,tol=0.1)
```

```python
model3.fit(xtrain,ytrain)
```

```python
model3.score(xtest,ytest)
```
