```python
nlp=spacy.load("en_core_web_sm")
doc=nlp("we just opened our wings, the flying part is coming soon")
for token in doc:
    if token.is_stop:
        print(token)
```

```python
def preprocess(text):
    doc=nlp(text)
    no_stop_words=[token.text for token in doc if not token.is_stop and not token.is_punct]
    return no_stop_words
```
